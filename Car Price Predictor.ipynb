```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('C:/Users/RESAF/Downloads/Data Science/Automobile_data.csv')
df.head()
```

```python
df.info()
```

```python
df.replace('?', np.nan, inplace = True)
```

```python
columns_to_convert = ['normalized-losses', 'bore', 'stroke','horsepower','peak-rpm','price']

df[columns_to_convert] = df[columns_to_convert].astype(float)
```

```python
df.info()
```

```python
df.shape
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.describe(include = 'object')
```

```python
df = df.drop(columns = ['engine-location','make'])
```

```python
df['num-of-cylinders'].unique()
```

```python
df['num-of-doors'].unique()
```

```python
df['num-of-cylinders'].replace({'four':4, 'five':5, 'six':6, 'three':3, 'eight':8},inplace = True)
df['num-of-doors'].replace({'two':2 , 'four':4},inplace = True)
```

```python
dummies = []
for column in df:
    if df[column].dtype == 'object':
        dummies.append(column)
dummies
```

```python
df1 = pd.get_dummies(data=df, columns = dummies)
df1.columns
```

```python
df1.head()
```

```python
cols_to_scale = ['symboling', 'normalized-losses', 'num-of-doors', 'wheel-base',
       'length', 'width', 'height', 'curb-weight', 'num-of-cylinders',
       'engine-size', 'bore', 'stroke', 'compression-ratio', 'horsepower',
       'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df1[cols_to_scale] = scaler.fit_transform(df1[cols_to_scale])
df1.sample(10)
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
x = df1.drop('price', axis=1)
y = df1['price']
```

```python
x = x.rename(columns=lambda x: x.replace('-', '_'))

# Updated column names
print(x.columns)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=10)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
models = [
    ('Linear Regression', LinearRegression(), {}),
    ('Decision Tree', DecisionTreeRegressor(), {'max_depth': [None, 5, 10]}),
    ('Random Forest', RandomForestRegressor(), {'n_estimators': [100,200,300]})
]
```

```python
best_model = None
best_score = float('-inf')
```

```python
for model_name, model, param_grid in models:
    grid_search = GridSearchCV(model, param_grid, cv=5)
    grid_search.fit(x_train, y_train)
    score = grid_search.best_score_
    if score > best_score:
        best_model = grid_search.best_estimator_
        best_score = score
    print(f'{model_name} Best score:', score)
    print(f'{model_name} Best parameters:', grid_search.best_params_)
    print()
```

```python
final_model = RandomForestRegressor(n_estimators = 300)
```

```python
final_model.fit(x_train,y_train)
```

```python
y_pred = final_model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Squared Error (MSE):", mse)
print("R-squared (R2):", r2)
```
